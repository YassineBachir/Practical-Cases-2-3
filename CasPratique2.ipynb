# Question 2:

```python
#Calcul de la moyenne
Tableau=Notes[Notes.columns[2:]].values
Tableau

```

```python
NotesPhysique=Tableau[:,0:3]
NotesPhysique
```

```python

Coef1=[2.5,2.5,1]
#M=[]
for j in range(len(NotesPhysique)):
    for i in range(0,3):
        NotesPhysique[j,i]=NotesPhysique[j,i]*Coef1[i]
        #M.append(M1)
NotesPhysique
moyen_Phy=np.sum(NotesPhysique,axis=1)/6

moyen_Phy

```

```python
#Validation 
V1=[]
credit_Phy=[]
for i in range(len(moyen_Phy)):
    if moyen_Phy[i] >=10:
        V1.append("V")
        credit_Phy.append(12)
        

    else:
        V1.append("NV")
        credit_Phy.append(0)
V1
    
```

# Moyenne Maths

```python

NotesMaths=Tableau[:,3:5]
NotesMaths
```

```python
#MATHS

Coef2=[2.5,2.5]
for j in range(len(NotesMaths)):
    for i in range(2):
        NotesMaths[j,i]=NotesMaths[j,i]*Coef2[i]
        #M.append(M1)
NotesMaths
moyen_Maths=np.sum(NotesMaths, axis=1)/5
moyen_Maths
```

```python
V2=[]
credit_Math=[]
for i in range(len(moyen_Maths)):
    if moyen_Maths[i] >=10:
        V2.append("V")
        credit_Math.append(10)
    else:
        V2.append("NV")
        credit_Math.append(0)
V2
```
